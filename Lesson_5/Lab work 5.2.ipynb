Створіть нову колонку "Actual price", яка утворюється із "Purchase Price" та націнки за принципом:
- якщо "Purchase Price" > 50, націнка 20%
- якщо "Purchase Price" > 100, націнка 10%
- в інших випадках націнка 30%
   

```python
def Actual_price(PurchasePrice):
    Act_PRICE=0
    if PurchasePrice > 50 and PurchasePrice<=100:
        Act_PRICE=PurchasePrice*1.2
    elif PurchasePrice > 100:
        Act_PRICE=PurchasePrice*1.1
    else:
        Act_PRICE=PurchasePrice*1.3
    return Act_PRICE


df['Actual price']=df["Purchase Price"].map(Actual_price)
df
```
