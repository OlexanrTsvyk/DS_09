### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import statistics
#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
#мінімальне та максимальне значення
def min_max(a):
    min_result = a[0]
    for val in a[1:]:
        if val < min_result:
            min_result = val
    max_result = a[0]
    for val in a[1:]:
        if val > max_result:
            max_result = val
    return min_result, max_result
print(min_max(list_with_random_numbers))

#середнє значення
def avg(a):
    sum1=0
    cnt=0
    for i in range (len(a)):
        sum1+=a[i]
        cnt+=1
    return sum1/cnt
print(avg(list_with_random_numbers))

#медіана послідовності
def med(a):
    b=sorted(a)
    c=0
    int1=len(b)//2
    if len(b)%2==0:
        c=(b[int1]+b[int1+1])/2
    else:
        c=b[int1+1]
    return c
print (med(list_with_random_numbers))  

#частоту кожної цифри в послідовності
def count_spisok(a):
    cnt=0
    set_1=set(a) #уникальные занчениея
    dict_1 = {}  # создаем пустой словарь
    for i in set_1:
        for j in a:
            if j==i:
                cnt+=1
        dict_1[i]=cnt
        cnt=0
    return dict_1
print (count_spisok(list_with_random_numbers))

#середнє квадратичне відхилення
def kvadrat_vidhulennya(a):
    sum1=0
    cnt=0
    sum_kvadrat=0
    for i in range (len(a)):
        sum1+=a[i]
        cnt+=1
    avg=sum1/cnt
    for i in a:
        sum_kvadrat+=(i-avg)**2
    return(((sum_kvadrat)**(1/2))/cnt)
print(kvadrat_vidhulennya(list_with_random_numbers))
```
